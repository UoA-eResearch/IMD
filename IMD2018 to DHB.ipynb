```python
import geopandas as gpd
import pandas as pd
pd.set_option('display.max_columns', None)
import os
os.listdir("input")
```

```python
imd = gpd.read_file("input/IMD2018.zip")
imd = imd.set_index("DZ2018")
imd
```

```python
for col in imd.columns[imd.columns.str.startswith("Rank_")]:
    imd[f"Census_Pop*{col}"] = imd.Census_Pop * imd[col]
imd = imd[list(imd.columns[imd.columns.str.startswith("Census_Pop")]) + ["geometry"]]
imd
```

```python
imd.geometry = imd.representative_point()
```

```python
DHB = gpd.read_file("input/statsnz-district-health-board-2015-SHP.zip")
DHB.to_crs(imd.crs, inplace=True)
DHB
```

```python
joined = gpd.sjoin(DHB, imd)
joined
```

```python
group = joined.groupby(joined.index)
```

```python
pop = group["Census_Pop"].sum()
```

```python
pop
```

```python
DHB["Census_Pop"] = pop
DHB
```

```python
DHB["Count_DZ2018"] = group.size()
```

```python
DHB.Count_DZ2018.value_counts()
```

```python
imd.columns[imd.columns.str.startswith("Census_Pop*")]
```

```python
for col in imd.columns[imd.columns.str.startswith("Census_Pop*")]:
    DHB[col.replace("Census_Pop*", "")] = group[col].sum() / pop
```

```python
DHB
```

```python
DHB.to_file("output/DHB_with_weighted_average_IMD18.shp.zip", driver='ESRI Shapefile')
```
