```python
import geopandas as gpd
import pandas as pd
pd.set_option('display.max_columns', None)
import os
os.listdir("input")
```

```python
imd = gpd.read_file("input/IMD2018.zip")
imd = imd.set_index("DZ2018")
imd
```

```python
for col in imd.columns[imd.columns.str.startswith("Rank_")]:
    imd[f"Census_Pop*{col}"] = imd.Census_Pop * imd[col]
imd = imd[list(imd.columns[imd.columns.str.startswith("Census_Pop")]) + ["geometry"]]
imd
```

```python
imd.geometry = imd.representative_point()
```

```python
sa2 = gpd.read_file("input/statsnz-statistical-area-2-2023-generalised-FGDB.zip!statistical-area-2-2023-generalised.gdb")
sa2 = sa2.set_index("SA22023_V1_00")
sa2
```

```python
assert sa2.crs == imd.crs
```

```python
joined = gpd.sjoin(sa2, imd)
joined
```

```python
group = joined.groupby(joined.index)
```

```python
pop = group["Census_Pop"].sum()
```

```python
pop
```

```python
sa2["Census_Pop"] = pop
sa2
```

```python
sa2["Count_DZ2018"] = group.size()
```

```python
sa2.Count_DZ2018.value_counts()
```

```python
imd.columns[imd.columns.str.startswith("Census_Pop*")]
```

```python
for col in imd.columns[imd.columns.str.startswith("Census_Pop*")]:
    sa2[col.replace("Census_Pop*", "")] = group[col].sum() / pop
```

```python
sa2
```

```python
sa2.to_file("output/SA2_2023_with_weighted_average_IMD18.shp.zip", driver='ESRI Shapefile')
```
